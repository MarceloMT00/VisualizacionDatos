```python
df = pd.read_csv('datasets/antropometria-dataset.csv')
df
```

```python
plt.figure()
plt.boxplot(df['peso'])
    
```

```python
tabla = pd.pivot_table(df,
                       index= [ 'desc_ent'],
                       columns = ['sexo'],
                       values = ['peso'],
                       aggfunc = {'peso':['median']}
                      )
tabla
```

```python
tabla.columns
```

```python
boxplot = tabla.boxplot(column=[('peso', 'median', 1)])
```

```python
tabla_2 = pd.pivot_table(df,
                          index=['desc_ent'],
                          columns=['sexo'],
                          values=['cintura'],
                          aggfunc = {'cintura':['median']}
                         )
tabla_2
```

```python
line_plot = tabla_2[('cintura','median', 2)].plot(kind='line')
```

```python
tabla_3 = pd.pivot_table(df,
                         index = ['desc_ent'],
                         columns = ['sexo'],
                         values = ['ropa'],
                         aggfunc = {'ropa':['sum']}
                        )
tabla_3
```

```python
roparo= tabla_3.sort_values(('ropa','sum',1),ascending=False)
top10 = roparo.head(10)
top10
```

```python
top10[('ropa','sum',1)].plot(kind='barh',
                             figsize = (30,20),
                             color = 'y',
                             alpha = 0.5)

```
