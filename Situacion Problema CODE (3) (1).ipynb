```python
import pandas as pd 
import numpy as np 
import matplotlib as mlp
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%ls
```

```python
df = pd.read_csv('wine-quality-white-and-red.csv')
df
```

```python
tabla_3 = pd.pivot_table(df,
                         index = ['type',],
                         columns = ['quality',],
                         values = ['alcohol',],
                         aggfunc = {'alcohol':['mean']}
                        )
tabla_3
```

```python
sns.lmplot(x='alcohol',
           y= 'quality',
           data=df,
           fit_reg= False,
           hue= 'type'
          )
plt.title('Scatterplot de calidad de alcohol vs calidad de vino')
```

```python
sns.swarmplot(x='alcohol',
              y='quality',
              data=df,
             )
plt.title('Swarmplot de alcohol vs quality del vino')
```

```python
df.plot.scatter(x= 'alcohol',
                y='quality');
```

# Regresion

```python
df.columns 
```

```python
model = sm.OLS.from_formula('alcohol~quality',
                            data=df)
result= model.fit()
result.summary()
```

```python
ax =sns.regplot(x='alcohol',
                y='quality',
                data=df,
                x_estimator = np.mean,
               )
```
