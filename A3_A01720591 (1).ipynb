```python
my_pivot = pd.pivot_table(df,
                          index = 'desc_ent',
                          values = 'peso',
                          aggfunc = ['mean','median'],
                         )
my_pivot
```

```python
df = pd.read_csv('datasets/adult.csv')
df.head()
```

```python
my_pivot = pd.pivot_table(df,
                          index = ['education'],
                          values = 'age',
                          aggfunc = ['mean'],
                          columns = 'marital-status',
                         )
my_pivot
```

```python
my_pivot = pd.pivot_table(df,
                          index = ['sex','education'],
                          values = 'age',
                          aggfunc = ['median'],
                          columns = 'marital-status',
                         )
my_pivot
```
